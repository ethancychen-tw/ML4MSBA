Below are what you will usually do:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
df = pd.read_csv("./house_price_train.csv")
cat_cols = list(df.select_dtypes("object").columns.values)
missing_val_cols = ["LotFrontage", "MasVnrArea", "GarageYrBlt"]
df.drop(cat_cols+missing_val_cols,axis='columns', inplace=True)
df.head()
````

````python
X = df.drop(['SalePrice'],axis = 'columns')
y = df['SalePrice']
````

````python
for col in X.columns.values:
    print(col)    
    sns.scatterplot(x=X[col], y=y)
    plt.show()
````

## Part1: Fitting a regression model for explanation
1. Choose X variables
2. View scatter plots (careful!)
3. Fit a regression model to data. Get estimated model
4. Check validity of model assumptions
5. Use estimated model to test/infer relationship in the population

````python
formula=f"SalePrice={' + '.join(X.columns.values)}"
print("Model: \n"+formula)
import statsmodels.api as sm
model = sm.OLS(y,X).fit()
model.summary()
````

And we need to check model assumption
error normality


````python
# residual plot
sns.scatterplot(model.resid)
plt.show()
model.resid.hist()
````

````python
# residual vs fitted
ax = sns.scatterplot(x=model.predict(X), y=model.resid)
ax.set(xlabel="fitted", ylabel="residual")
````

````python
# residual vs variable
for col in X.columns:
    sns.scatterplot(x=X[col], y=model.resid)
    plt.show()
````

- Is the coef significant? What does this mean?
- What does the $R^2$ mean?

````python
print("SalePrice = "+" + ".join([f"{round(coef,2)} {ind} "for ind, coef in model.params.items()]))
````

- Is $R^2$ useful for evaluating prediction quality?

- General evaluation method:
1. Partition data into training and validation set
2. Fit model to training set
3. Evaluate predictions on validation set

- How to evaluate prediction?

Mean absolute error (MAE),  Mean Squared Error (MSE), Mean absolute % error (MAPE)

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
````

````python
model = LinearRegression().fit(X_train, y_train)

from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
y_pred_train = model.predict(X_train)
y_pred_test = model.predict(X_test)
print("----")
print("train data")
print(f"MAE: {mean_absolute_error( y_train, y_pred_train)}\nRMSE: {mean_squared_error(y_train, y_pred_train)**(1/2)} \nMAPE:{mean_absolute_percentage_error(y_train, y_pred_train)}")
print("----")
print("test data")
print(f"MAE: {mean_absolute_error( y_test, y_pred_test)}\nRMSE: {mean_squared_error(y_test, y_pred_test)**(1/2)} \nMAPE:{mean_absolute_percentage_error(y_test, y_pred_test)}")
````

- How to intepret this result?

Let's fit another model. This time we fit a model without `MoSold` and `YrSold`. 


````python
X_train2 = X_train.drop(['MoSold', 'YrSold'], axis='columns')
X_test2 = X_test.drop(['MoSold', 'YrSold'], axis='columns')
model2 = LinearRegression().fit(X_train2, y_train)

y_pred_train2 = model2.predict(X_train2)
y_pred_test2 = model2.predict(X_test2)
print("----")
print("train data")
print(f"MAE: {mean_absolute_error( y_train, y_pred_train2)}\nRMSE: {mean_squared_error(y_train, y_pred_train2)**(1/2)} \nMAPE:{mean_absolute_percentage_error(y_train, y_pred_train2)}")
print("----")
print("test data")
print(f"MAE: {mean_absolute_error( y_test, y_pred_test2)}\nRMSE: {mean_squared_error(y_test, y_pred_test2)**(1/2)} \nMAPE:{mean_absolute_percentage_error(y_test, y_pred_test2)}")
````
